# **Introdução**

A energia elétrica constitui um insumo fundamental para o desenvolvimento econômico e social, estando diretamente relacionada à qualidade de vida da população e ao funcionamento das atividades produtivas. No Brasil, o consumo de energia elétrica tem apresentado crescimento contínuo nas últimas décadas, acompanhando a expansão econômica e demográfica do país. Segundo dados oficiais do Ministério de Minas e Energia e da Empresa de Pesquisa Energética, o consumo nacional ultrapassa 550 TWh anuais, sendo o setor residencial responsável por cerca de 30% da demanda total.

Apesar desse crescimento, o consumo energético apresenta desigualdades regionais relevantes, influenciadas por fatores socioeconômicos, climáticos e estruturais. A região Nordeste responde por aproximadamente 18% da demanda elétrica brasileira e apresenta heterogeneidades internas significativas entre seus estados, refletindo diferenças de renda, condições ambientais e dinâmica econômica. Nesse contexto, análises baseadas em médias regionais tendem a ocultar disparidades intrarregionais e dificultar a identificação de situações de vulnerabilidade energética.

Diante desse cenário, este projeto insere-se na área de Ciência de Dados aplicada ao planejamento energético, propondo a integração de séries temporais de consumo de energia elétrica com indicadores socioeconômicos por meio de técnicas de aprendizado de máquina. O estudo busca analisar padrões regionais de consumo e desenvolver modelos preditivos capazes de antecipar variações na demanda energética, contribuindo para o planejamento eficiente do setor e para a formulação de políticas públicas orientadas à redução das desigualdades no acesso e no uso da energia elétrica.

Portanto, o que nos moveu foi o desafio de transformar essa massa de dados em um termômetro social, capaz de antecipar problemas e garantir que a infraestrutura e o consumo caminhem juntos com a proteção da dignidade humana.

# **Descrição da base de dados**

Neste projeto, vamos utilizar conjuntos de dados estruturados que apresentam o consumo mensal de energia elétrica no período de janeiro de 2015 a dezembro de 2025. O primeiro dataset reúne dados do consumo das cinco regiões do Brasil, permitindo uma análise macro do comportamento do consumo no país. Complementarmente, o segundo dataset — derivado do primeiro através de um recorte geográfico — focaliza exclusivamente os nove estados do Nordeste, possibilitando uma análise detalhada das sazonalidades e particularidades socioeconômicas locais.

Os dados estão organizados em formato "long" (ou Tidy Data), onde cada linha representa uma observação única de consumo para uma determinada combinação de tempo, local e categoria. Essa organização é ideal para o treinamento de modelos de redes neurais como a LSTM, pois facilita a criação de janelas deslizantes temporais.

A forma dos dados é tabular, processada inicialmente como um DataFrame no Pandas e exportada para arquivos .csv. O período de coleta abrange 11 anos de histórico mensal, totalizando uma profundidade temporal de 132 meses por série, o que confere robustez estatística para a identificação de tendências e ciclos sazonais.

Os dados brutos provêm do Ministério de Minas e Energia (MME) e da Empresa de Pesquisa Energética (EPE), entidades responsáveis pelo levantamento oficial do consumo de energia elétrica faturada por classe de consumo no Brasil. O acesso foi realizado pela Base dos Dados (BD), utilizando a biblioteca **basedosdados** em Python para realizar consultas SQL através de Google BigQuery no dataset **br_mme_consumo_energia_eletrica**.

Considerando que este projeto deve ser desenvolvido com foco nas ODS 9 e ODS 12, as variáveis selecionadas foram:

* **ano / mes**: Dimensões temporais para a série.

* **regiao / sigla_uf**: Dimensões espaciais para comparação.

* **tipo_consumo**: Categorias do tipo de consumo (Industrial, Comercial, Residencial, etc.).

* **consumo**: O volume de energia faturada em megawatt-hora no mês

O dataset com os dados de consumo de energia segmentado por região do Brasil apresenta 3.240 obervações, enquanto a base de dados focada no consumo dos estados nordestinos apresenta 5.832 observações.
