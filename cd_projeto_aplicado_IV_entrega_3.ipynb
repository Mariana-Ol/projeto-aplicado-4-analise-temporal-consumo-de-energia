# **Referencial Teórico**

O consumo de energia elétrica é um importante indicador do desenvolvimento econômico e social, estando diretamente associado à renda, urbanização e condições climáticas de uma região (EPE, 2024). Entretanto, esse consumo apresenta variações espaciais e temporais relevantes, o que torna necessária a utilização de métodos analíticos capazes de capturar padrões ao longo do tempo e entre diferentes localidades.

Nesse contexto, a análise de séries temporais se destaca como uma abordagem fundamental, pois permite modelar dados observados em intervalos regulares e identificar componentes estruturais como tendência, sazonalidade e ruído (HYNDMAN; ATHANASOPOULOS, 2021). A decomposição dessas séries facilita a compreensão do comportamento do consumo energético, evidenciando padrões recorrentes e variações de longo prazo.

Em dados de consumo energético, é comum observar comportamentos sazonais associados a fatores climáticos, além de tendências de crescimento relacionadas ao desenvolvimento econômico. Modelos de decomposição podem ser classificados como aditivos ou multiplicativos, sendo este último frequentemente aplicado quando a variação sazonal cresce proporcionalmente à tendência, característica comum em séries desse tipo (HYNDMAN; ATHANASOPOULOS, 2021).

Tradicionalmente, modelos estatísticos como o ARIMA são utilizados para previsão de séries temporais, apresentando boa interpretabilidade e desempenho em séries lineares (BOX; JENKINS; REINSEL, 2015). No entanto, esses modelos possuem limitações na captura de padrões não lineares e de longo prazo. Como alternativa, técnicas de machine learning e deep learning, especialmente redes neurais recorrentes do tipo LSTM, têm sido amplamente aplicadas na previsão de demanda energética, devido à sua capacidade de aprender dependências temporais complexas (HOCHREITER; SCHMIDHUBER, 1997).

Outra abordagem relevante consiste na integração de dados energéticos com indicadores socioeconômicos, como renda e PIB, permitindo análises mais abrangentes sobre o consumo e possibilitando a identificação de situações de vulnerabilidade energética. Essa perspectiva está alinhada aos princípios da Ciência de Dados, que busca extrair conhecimento a partir da combinação de múltiplas fontes de dados (PROVOST; FAWCETT, 2013).

Trabalhos correlacionados apontam que análises agregadas por grandes regiões podem ocultar desigualdades internas, tornando essencial o uso de dados desagregados por estado ou localidade (EPE, 2023). Nesse sentido, a abordagem adotada neste projeto — que combina análise exploratória, modelagem preditiva e integração de dados socioeconômicos — permite uma compreensão mais detalhada das disparidades regionais.

Dessa forma, a solução proposta integra técnicas estatísticas e de aprendizado de máquina para não apenas prever o consumo de energia, mas também interpretar seus determinantes, contribuindo para um planejamento energético mais eficiente e para a formulação de políticas públicas alinhadas aos Objetivos de Desenvolvimento Sustentável (ODS) 9 e 12.

O pipeline da solução proposta está estruturado em etapas sequenciais que integram coleta, tratamento, análise e modelagem dos dados, com o objetivo de gerar insights sobre o consumo de energia elétrica no Brasil.

Em seguida, realiza-se o pré-processamento dos dados, incluindo a verificação da consistência e padronização das variáveis categóricas, análise de valores nulos e conversão da variável de data para o formato adequado (datetime), garantindo a correta aplicação de técnicas de séries temporais.

Posteriormente, é realizada a modelagem preditiva, iniciando com a preparação dos dados para séries temporais. Na fase de avaliação do modelo, são utilizadas métricas como MAE, MAPE e RMSE para medir o desempenho das previsões, além da comparação entre os modelos desenvolvidos, permitindo identificar a abordagem mais adequada.

Por fim, na etapa de interpretação dos resultados, os resultados são analisados para identificar padrões de vulnerabilidade energética, avaliar o impacto da renda no consumo de energia e compreender as diferenças regionais, fornecendo subsídios para a tomada de decisão e formulação de políticas públicas.

## Análise dos componentes das séries temporais

### Decomposição da região Nordeste

```python
# Criando um filtro para isolar a série que queremos analisar
filtro = (df_regioes['regiao'] == 'Nordeste') & (df_regioes['tipo_consumo'] == 'Residencial')
serie_nordeste_res = df_regioes[filtro]['consumo']


print(serie_nordeste_res.head())
```

```python
resultado = seasonal_decompose(serie_nordeste_res, model='multiplicative', period=12)

plt.rcParams.update({'figure.figsize': (10,8)})
resultado.plot()
plt.show()
```

### Como a sazonalidade não ficou muito clara no gráfico, vamos visualizá-la com uma "lupa" do código a seguir:

```python
import matplotlib.dates as mdates

#Seleciona apenas os primeiros 24 meses (2 anos) para o gráfico não ficar apertado
sazonal_zoom = resultado.seasonal.iloc[:24]

plt.figure(figsize=(12, 4))
sazonal_zoom.plot(marker='o', color='green', linewidth=2)

# Formata o eixo X para mostrar o nome dos meses
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b'))

plt.title('Zoom da sazonalidade (NE)')
plt.grid(True, alpha=0.3)
plt.show()
```

## Principais obervações:

1 - Há uma tendência de crescimento no consumo de energia residencial na região NE, o que se observa no gráfico da tendência.

2 - O gráfico da sazonalidade revela que o consumo de energia residencial tende a aumentar a partir do mês de fevereiro, atingindo seu pico no mês de outubro. O nível do consumo de energia apresenta aumento mais acentuado a partir de julho, e começa a decair em outubro.

3 - Identicamos um vale profundo no mês de fevereiro, o que é contraintuitivo, pois este é um dos meses mais quentes em todo o país. Uma possível explicação para este fenômeno é que seja um reflexo do menor número de dias desse mês em comparação aos demais, o que reduz o volume total de energia contabilizado (efeito calendário).

4 - De setembro a dezembro, o Nordeste sofre o efeito B-R-Ó BRÓ (UOL, 2024), fenômeno climático que provoca calor extremo da região. As consequências deste fenômeno regional se refletem neste gráfico da sazonalidade, uma vez que os meses do B-R-Ó BRÓ apresentam um consumo elevado de energia residencial.

5 - O gráfico de resíduos apresenta-se estabilizado em torno de 1.0, indicando que o modelo multiplicativo foi capaz de capturar a maior parte da variabilidade da série. Pequenas flutuações residuais observadas sugerem eventos estocásticos (como variações climáticas extremas ou mudanças tarifárias pontuais) que não seguem o padrão sazonal típico.

Essa trajetória sugere que esses dois estados foram menos impactados pelos efeitos da crise hídrica no consumo doméstico em comparação aos demais estados da região, mantendo a tendência de alta mesmo nos anos de maior estresse hídrico e tarifário.

Por fim, observa-se que, apesar dos percalços como períodos de crise hídrica, a tendência geral da região ao longo dos anos de 2015 a 2023 é de crescimento no consumo energético residencial.

# **Modelo base**

Aplicação de, pelo menos, um primeiro modelo (modelo de aprendizado de máquina ou modelo estatístico) que deve ser refinado até a entrega final do projeto. Apresente o método aplicado e a análise dos resultados obtidos.

Para a nossa investigação, adotamos o SARIMA (Seasonal Autoregressive Integrated Moving Average) como modelo base, uma técnica estatística clássica que se destaca pela capacidade de capturar, em uma única estrutura, a tendência de crescimento, a dependência temporal e as oscilações sazonais do consumo. A escolha foi inspirada em discussões recentes sobre o planejamento energético no subsistema Nordeste (COELHO et al., 2026), que reforçam como, em séries altamente agregadas com padrões estruturais estáveis, modelos estatísticos tradicionais podem muitas vezes superar abordagens de maior complexidade computacional (como redes neurais), oferecendo uma previsão mais confiável, interpretável e com menor custo de processamento.
Para aplicar o modelo, focamos o recorte inicial em Maranhão, Ceará e Bahia, pois esses estados cenários contrastantes, como constatado em nossa análise exploratória: a alta elasticidade-renda maranhense, a dependência climática cearense durante o período crítico do "B-R-O-Bró" e a estabilidade rígida do consumo baiano. Esse trio funciona como um bom cenário de teste, garantindo que o modelo seja capaz de processar realidades onde o consumo de energia varia de um bem de conforto a uma necessidade vital de subsistência, pavimentando o caminho para as simulações preditivas que faremos nas próximas semanas.

No modelo-base, usamos apenas os dados de consumo nesta fase inicial para criar um baseline (uma linha de base), capturando como a tendência e a sazonalidade se comportam isoladamente ao longo dos anos. É o modelo que nos diz: "se nada mudar e a economia ficar parada, o consumo deve seguir este padrão".

NOs próximos passos, utilizaremos o modelo SARIMAX, para podermos incluir a variável socioenconômida RENDA. Como o artigo que estamos usando como base aponta (COELHO et al., 2026), a ausência de indicadores econômicos é uma limitação importante, pois eles ajudam a explicar flutuações que a série de consumo sozinha não consegue justificar. Assim, evoluir para o SARIMAX nos permitirá cruzar a necessidade climática com a capacidade financeira, refinando a previsão para cenários de crise ou crescimento econômico.

```python
"""Usamos .asfreq('MS') para garantir que o Python sabe que os dados são mensais
Definimos a data como o índice. Isto é vital para o SARIMA, pois ele precisa
de "olhar para o calendário" para entender os ciclos de cada estado."""
serie_ma = df_residencial[df_residencial['sigla_uf'] == 'MA'].set_index('data')['consumo'].asfreq('MS')
serie_ce = df_residencial[df_residencial['sigla_uf'] == 'CE'].set_index('data')['consumo'].asfreq('MS')
serie_ba = df_residencial[df_residencial['sigla_uf'] == 'BA'].set_index('data')['consumo'].asfreq('MS')

# Conferir se os dados estão corretos (devem aparecer as datas e os valores)
print("Série Maranhão (primeiras linhas):")
print(serie_ma.head())

print("\nSérie Ceará (primeiras linhas):")
print(serie_ce.head())

print("\nSérie Bahia (primeiras linhas):")
print(serie_ba.head())
```

O Teste de Dickey-Fuller Aumentado (ADF) serve para verificar se a série temporal é estacionária. Em termos simples, uma série estacionária é aquela que se comporta de forma estável ao longo do tempo, sem uma tendência clara de subida ou descida e com oscilações constantes.

Para o modelo SARIMA funcionar bem, precisamos saber se os dados são estáveis. Se não forem, o modelo terá de aplicar uma "diferenciação" (o parâmetro d) para estabilizar os números.

```python
from statsmodels.tsa.stattools import adfuller

def realizar_teste_adf(serie, nome_estado):
    print(f"--- Teste ADF: {nome_estado} ---")

    resultado = adfuller(serie.dropna())

    p_valor = resultado[1]
    print(f"p-valor: {p_valor:.4f}")

    if p_valor <= 0.05:
        print("A série é ESTACIONÁRIA.")
    else:
        print("A série NÃO é estacionária.")
    print("-" * 30)

# Executar para os três estados
realizar_teste_adf(serie_ma, "Maranhão")
realizar_teste_adf(serie_ce, "Ceará")
realizar_teste_adf(serie_ba, "Bahia")
```

Se o p-valor for menor ou igual a 0.05, a série é estável.

Se o p-valor for maior que 0.05, a série tem uma tendência (provavelmente de subida, como é comum no consumo de energia) e não é estacionária.

### Construção do modelo baseline


Separar os dados em conjunto de treino (2015 até 2022) e conjunto de teste (o ano de 2023).

Configurar o modelo com $d=1$ (para tratar a não estacionaridade) e $s=12$ (para a sazonalidade mensal).

Calcular as métricas de desempenho.

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

def treinar_baseline_sarima(serie, nome_estado):
    print(f"\n--- Treinando Baseline: {nome_estado} ---")

    # 1. Divisão Treino (até 2022) e Teste (2023)
    treino = serie[:'2022-12-01']
    teste = serie['2023-01-01':]

    # 2. Configuração do Modelo SARIMA (1,1,1)x(1,1,1,12)
    # O d=1 aqui é o que trata a não-estacionaridade que você encontrou!
    modelo = SARIMAX(treino,
                     order=(1, 1, 1),
                     seasonal_order=(1, 1, 1, 12),
                     enforce_stationarity=False,
                     enforce_invertibility=False)

    resultado = modelo.fit(disp=False)

    # 3. Previsão para os 12 meses de 2023
    previsao = resultado.get_forecast(steps=12).predicted_mean

    # 4. Cálculo das métricas
    mae = mean_absolute_error(teste, previsao)
    mape = np.mean(np.abs((teste - previsao) / teste)) * 100

    print(f"MAE: {mae:.2f} MWh")
    print(f"MAPE: {mape:.2f}%")

    return previsao, teste, mape

# Agora vamos rodar para os três estados e guardar os resultados
prev_ma, teste_ma, mape_ma = treinar_baseline_sarima(serie_ma, "Maranhão")
prev_ce, teste_ce, mape_ce = treinar_baseline_sarima(serie_ce, "Ceará")
prev_ba, teste_ba, mape_ba = treinar_baseline_sarima(serie_ba, "Bahia")
```

O Maranhão (4,18%) e a Bahia (4,25%) tiveram erros muito parecidos, enquanto o Ceará (5,78%) teve um erro um pouco maior.Aparentemente, o Ceará tem oscilações que o modelo de energia sozinho tem mais dificuldade de prever.

Os modelos de referência apresentaram um erro percentual médio (MAPE) inferior a 6%, validando a escolha da arquitetura SARIMA para a representação da dinâmica energética regional, conforme observado em estudos similares na literatura (Eshily et al., 2026).
