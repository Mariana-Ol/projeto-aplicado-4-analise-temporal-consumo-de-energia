# **Introdução**

A energia elétrica é um recurso essencial para o desenvolvimento econômico e social, diretamente relacionada à qualidade de vida e ao funcionamento das atividades produtivas. No entanto, esse consumo apresenta desigualdades regionais relevantes, especialmente na região Nordeste, onde diferenças socioeconômicas e climáticas geram disparidades que muitas vezes são ocultadas por análises baseadas em médias.

Diante desse contexto, este projeto tem como objetivo geral desenvolver uma solução baseada em análise de dados e modelos preditivos para apoiar a gestão eficiente do consumo de energia elétrica no Brasil, contribuindo para a tomada de decisões e para a implementação de políticas públicas voltadas à redução da vulnerabilidade energética.

Como objetivos específicos, destacam-se: a coleta e tratamento de dados de consumo energético (MME/EPE); a análise de padrões regionais, com foco no Nordeste; a investigação da relação entre consumo e indicadores socioeconômicos, como PIB e renda; o desenvolvimento de um modelo preditivo utilizando redes neurais LSTM; e a proposição de diretrizes alinhadas aos Objetivos de Desenvolvimento Sustentável (ODS) 9 e 12.

Nesse sentido, o projeto se justifica pela aplicação de técnicas de machine learning para transformar dados históricos em inteligência analítica, permitindo identificar padrões, antecipar demandas e apoiar um planejamento energético mais eficiente e equitativo.

# **Referencial Teórico**

O consumo de energia elétrica é um importante indicador do desenvolvimento econômico e social, estando diretamente associado à renda, urbanização e condições climáticas de uma região (EPE, 2024). Entretanto, esse consumo apresenta variações espaciais e temporais relevantes, o que torna necessária a utilização de métodos analíticos capazes de capturar padrões ao longo do tempo e entre diferentes localidades.

Nesse contexto, a análise de séries temporais se destaca como uma abordagem fundamental, pois permite modelar dados observados em intervalos regulares e identificar componentes estruturais como tendência, sazonalidade e ruído (HYNDMAN; ATHANASOPOULOS, 2021). A decomposição dessas séries facilita a compreensão do comportamento do consumo energético, evidenciando padrões recorrentes e variações de longo prazo.

A análise exploratória realizada neste projeto evidenciou a presença de tendência de crescimento no consumo residencial em todas as regiões do Brasil, além de padrões sazonais distintos. No Nordeste, por exemplo, observou-se aumento significativo do consumo nos meses mais quentes, especialmente entre setembro e dezembro, possivelmente associado ao fenômeno climático conhecido como B-R-O-Bró (UOL, 2024). Já no Sudeste, os picos de consumo estão mais relacionados a períodos de calor no início do ano. Esses resultados reforçam a influência de fatores climáticos e regionais no comportamento do consumo energético.

Além disso, a decomposição das séries temporais indicou que o modelo multiplicativo é adequado para esse tipo de dado, uma vez que a amplitude das variações sazonais cresce proporcionalmente à tendência, característica comum em séries de consumo energético (HYNDMAN; ATHANASOPOULOS, 2021). A análise dos resíduos demonstrou distribuição estável, sugerindo que os principais padrões foram capturados adequadamente.

Tradicionalmente, modelos estatísticos como o ARIMA são utilizados para previsão de séries temporais, apresentando boa interpretabilidade e desempenho em séries lineares (BOX; JENKINS; REINSEL, 2015). No entanto, esses modelos possuem limitações na captura de padrões não lineares e de longo prazo. Como alternativa, técnicas de machine learning e deep learning, especialmente redes neurais recorrentes do tipo LSTM, têm sido amplamente aplicadas na previsão de demanda energética, devido à sua capacidade de aprender dependências temporais complexas (HOCHREITER; SCHMIDHUBER, 1997).

Outra abordagem relevante consiste na integração de dados energéticos com indicadores socioeconômicos, como renda e PIB, permitindo análises mais abrangentes sobre o consumo e possibilitando a identificação de situações de vulnerabilidade energética. Essa perspectiva está alinhada aos princípios da Ciência de Dados, que busca extrair conhecimento a partir da combinação de múltiplas fontes de dados (PROVOST; FAWCETT, 2013).

Trabalhos correlacionados apontam que análises agregadas por grandes regiões podem ocultar desigualdades internas, tornando essencial o uso de dados desagregados por estado ou localidade (EPE, 2023). Nesse sentido, a abordagem adotada neste projeto — que combina análise exploratória, decomposição de séries temporais e modelagem preditiva — permite uma compreensão mais detalhada das disparidades regionais.

Dessa forma, a solução proposta integra técnicas estatísticas e de aprendizado de máquina para não apenas prever o consumo de energia, mas também interpretar seus determinantes, contribuindo para um planejamento energético mais eficiente e para a formulação de políticas públicas alinhadas aos Objetivos de Desenvolvimento Sustentável (ODS) 9 e 12.
